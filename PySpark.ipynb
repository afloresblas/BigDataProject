# Pre - Procesamiento (Limpieza)

##### Número de filas Únicos y Duplicados

```
%time

# Valores totales
print("Valores totales", df_application.count())
# Valores únicos
print("Valores únicos", df_application.select("ID").distinct().count())
```

```
%time

# Valores totales
print("Valores totales", df_credit.count())
# Valores únicos
print("Valores únicos", df_credit.select("ID").distinct().count())
```

##### Análisis de valores duplicados df_application
Los valores con ID duplicado del registro de aplicaciones serán eliminados

```
%time

df_application = df_application.dropDuplicates(['ID'])
```

```
%time

# Valores totales
print("Valores totales", df_application.count())
# Valores únicos
print("Valores únicos", df_application.select("ID").distinct().count())
```

##### Número de id's que las dos tablas comparten

```
%time

df_application.select("ID").intersect(df_credit.select("ID")).count()
```

##### Análisis de valores nulos df_application

```
%time

df_application.select([F.count(F.when(F.col(c).isNull(),c)).alias(c) for c in df_application.columns]).display()
```

##### Campo OCCUPATION_TYPE (df_application) 
Tiene gran cantidad de valores nulos --> No se tomará en cuenta este campo

```
%time

df_application = df_application.drop('OCCUPATION_TYPE') 
```

```
%time

df_application.select([F.count(F.when(F.col(c).isNull(),c)).alias(c) for c in df_application.columns]).display()
```

##### Análisis de valores nulos df_credit
No hay valores nulos

```
%time

df_credit.select([F.count(F.when(F.col(c).isNull(),c)).alias(c) for c in df_credit.columns]).display()
```

##### Unificar los 2 datasets

```
%time

df_union = df_application.join(df_credit, 'ID', how = 'inner')
```

##### Eliminar las columnas que no se usarán para el análisis

```
%time

df_union = df_union.drop('FLAG_MOBIL', 'FLAG_WORK_PHONE', 'FLAG_PHONE', 'FLAG_EMAIL') 
```

```
%time

df_union.limit(5).display()
```

### Cast de datos y eliminación de filas que contengan valores nulo

```
%time

df_union_1['AMT_INCOME_TOTAL'] = df_union_1['AMT_INCOME_TOTAL'].astype('float').astype('int')
df_union_2 = df_union_1.dropna()
```

```
%time

print("Las familias que tienen dos hijos son las que mayor ingreso perciben")
df_union_2.groupby(['AMT_INCOME_TOTAL', 'CNT_FAM_MEMBERS']).size().plot(kind='barh', figsize=(80, 80), grid=True)
```

```
%time

le = LabelEncoder()
for x in df_union_2:
    if df_union_2[x].dtypes=='object':
        df_union_2[x] = le.fit_transform(df_union_2[x])
```

```
%time

fig, ax= plt.subplots(nrows= 2, ncols = 3, figsize= (14,6))
sns.scatterplot(x='ID', y='CNT_CHILDREN', data=df_union_2, ax=ax[0][0], color= 'orange')
sns.scatterplot(x='ID', y='AMT_INCOME_TOTAL', data=df_union_2, ax=ax[0][1], color='orange')
sns.scatterplot(x='ID', y='DAYS_BIRTH', data=df_union_2, ax=ax[0][2])
sns.scatterplot(x='ID', y='DAYS_EMPLOYED', data=df_union_2, ax=ax[1][0])
sns.scatterplot(x='ID', y='CNT_FAM_MEMBERS', data=df_union_2, ax=ax[1][1], color= 'orange')
```

#### Existen valores atípicos en el campo CNT_CHILDREN

```
%time


upper_limit = df_union_2['CNT_CHILDREN'].quantile(0.999)
lower_limit = df_union_2['CNT_CHILDREN'].quantile(0.001)
df_union_2 = df_union_2[(df_union_2['CNT_CHILDREN']>lower_limit) & (df_union_2['CNT_CHILDREN']<upper_limit)]
```

#### Existen valores atípicos en el campo AMT_INCOME_TOTAL

```
%time 

upper_limit = df_union_2['AMT_INCOME_TOTAL'].quantile(0.999)
lower_limit = df_union_2['AMT_INCOME_TOTAL'].quantile(0.001)
df_union_2= df_union_2[(df_union_2['AMT_INCOME_TOTAL']>lower_limit) & (df_union_2['AMT_INCOME_TOTAL']<upper_limit)]
```

```
%time

fig, ax= plt.subplots(nrows= 2, ncols = 3, figsize= (14,6))
sns.scatterplot(x='ID', y='CNT_CHILDREN', data=df_union_2, ax=ax[0][0], color= 'orange')
sns.scatterplot(x='ID', y='AMT_INCOME_TOTAL', data=df_union_2, ax=ax[0][1], color='orange')
sns.scatterplot(x='ID', y='DAYS_BIRTH', data=df_union_2, ax=ax[0][2])
sns.scatterplot(x='ID', y='DAYS_EMPLOYED', data=df_union_2, ax=ax[1][0])
sns.scatterplot(x='ID', y='CNT_FAM_MEMBERS', data=df_union_2, ax=ax[1][1], color= 'orange')
```
